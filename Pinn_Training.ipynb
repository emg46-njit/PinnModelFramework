```python
iterations = 8000 #@param {type:"number"}

data_config = {
  'type': 'fits',
  'slices': [{
      'fits_path' : {
        'Br': br_path,
        'Bt': bt_path,
        'Bp': bp_path
      }
  }],
  'iterations': iterations
}
```

```python
#@title Logging Settings
entity = "emg46-new-jersey-institute-of-technology" #@param {type:"string"}
project = "PINN modeling" #@param {type:"string"}
joinedname = "Div Free" + ": AR" +  ARLabel + ": " + version
name = joinedname #@param {type:"string"}
print(joinedname)
key = "a06d8938922438464cb52cc11bf67943d9da181c" #@param {type:"string"}
continue_training = False #@param {type:"boolean"}
id = "" #@param {type:"string"}

entity = None if entity == "" else entity
project = None if project == "" else project
logging_config = {'log_model': 'all', 'entity': entity, 'project': project, 'name': name}

if continue_training:
  logging_config['id'] = id
```

```python
#@title Advanced model settings
#@markdown Keep these two settings true, required for DIV-free models; custom settings allows modification of weight strength
use_custom_settings = True #@param {type:"boolean"}
vector_potential = True #@param {type:"boolean"}
#@markdown Epoch Count
dim = 64 #@param {type:"number"}

if use_custom_settings:
  model_config = {'type': 'vector_potential' if vector_potential else 'b',
                  'dim': dim}
else:
  model_config = None
```

```python
#@title Advanced training settings
use_custom_settings = True #@param {type:"boolean"}
#@markdown Force-Free Condition loss Weight
lambda_ff = 0.1 #@param {type:"number"}
#@markdown Div-Free Condition loss Weight
lambda_div = 0.1 #@param {type:"number"}
#@markdown Boundary Condition loss Weight
lambda_boundary = 0.5 #@param {type:"number"}
#@markdown Epoch Count
epochs = 10 #@param {type:"number"}



if use_custom_settings:
  boundary_loss = {'type': 'boundary', 'name': 'boundary', 'lambda': lambda_boundary,
                 'ds_id': ['boundary_01', 'potential']}
  force_free_loss = {'type': 'force_free', 'lambda': lambda_ff}
  divergence_free_loss = {'type': 'divergence', 'lambda': lambda_div}

  training_config = {'loss_config': [boundary_loss, force_free_loss, divergence_free_loss], 'epochs': epochs}
else:
  training_config = {}
```

```python
import wandb

wandb.login(key=key)

run(download_dir_extrapolation, data=data_config,
    logging=logging_config,
    training=training_config,
    model=model_config)
```

```python
#@title Convert to NPY

from nf2.evaluation.output import CartesianOutput

nf2_path = os.path.join(download_dir_extrapolation, 'extrapolation_result.nf2')
mmpixels = 0.72 #@param {type: "number"}

nf2_out = CartesianOutput(nf2_path)
output = nf2_out.load_cube(Mm_per_pixel=mmpixels, progress=True)

print('NF2 Outputs:')
for k, v in output.items():
  print(f'{k} : {v.shape if hasattr(v, "shape") else v}')

converted_path = os.path.join(download_dir_extrapolation, "Converted Data")
os.makedirs(converted_path, exist_ok=True)

b_npy = os.path.join(converted_path, 'MagneticField.npy')
a_npy = os.path.join(converted_path, 'MagneticVectorPotential.npy')
jac_npy = os.path.join(converted_path, 'JacobianMatrix.npy')
coords_npy = os.path.join(converted_path, 'Coords.npy')
mm_pixels_txt = os.path.join(converted_path, 'MMPerPixel.txt')

np.save(b_npy, output['b'].value)
np.save(a_npy, output['a'].value)
np.save(jac_npy, output['jac_matrix'].value)
np.save(coords_npy, output['coords'])

with open(mm_pixels_txt, 'w') as file:
    file.write(f"Megameters Per Pixel: {output['Mm_per_pixel']}")
```
